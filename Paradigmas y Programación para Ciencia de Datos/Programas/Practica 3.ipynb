```python
import math

def cifrar(texto, clave):
    """
    Cifra un texto utilizando el método de transposición columnar con clave.

    Args:
        texto: El mensaje que se va a cifrar.
        clave: La palabra clave para el cifrado.

    Returns:
        El texto cifrado.
    """



    texto_plano = texto.replace(" ", "").upper()
    num_columnas = len(clave)
    num_filas = math.ceil(len(texto_plano) / num_columnas)
    caracteres_faltantes = (num_filas * num_columnas) - len(texto_plano)
    texto_plano += "X" * caracteres_faltantes

    indices_clave = sorted([(letra, i) for i, letra in enumerate(clave)])


    texto_cifrado = ""

    for letra, indice_columna in indices_clave:
        columna = ""

        for i in range(num_filas):
            puntero = indice_columna + (i * num_columnas)
            columna += texto_plano[puntero]
        texto_cifrado += columna

    return texto_cifrado




def descifrar(texto_cifrado, clave):
    """
    Descifra un texto utilizando el método de transposición columnar con clave.

    Args:
        texto_cifrado: El mensaje cifrado.
        clave: La palabra clave usada para el cifrado.

    Returns:
        El texto descifrado.
    """

    num_columnas = len(clave)
    num_filas = math.ceil(len(texto_cifrado) / num_columnas)

    indices_clave_ordenada = sorted([(letra, i) for i, letra in enumerate(clave)])

    tabla_descifrado = [[""] * num_columnas for _ in range(num_filas)]

    puntero = 0
    for _, indice_columna in indices_clave_ordenada:
        for i in range(num_filas):
            tabla_descifrado[i][indice_columna] = texto_cifrado[puntero]
            puntero += 1


    texto_plano = ""
    for fila in tabla_descifrado:
        texto_plano += "".join(fila)

    return texto_plano.rstrip("X")



texto_original = "El clima de hoy 22 de septiempre  en Coyoacán (Ciudad de México) será mayormente soleado por la mañana con temperaturas alrededor de 22°C, y se espera un aumento en la nubosidad y una ligera probabilidad de lluvias para la tarde y noche"
clave = "Clima"

print(texto_original)
print(cifrar(texto_original, clave))
print(descifrar(cifrar(texto_original, clave), clave))
```
