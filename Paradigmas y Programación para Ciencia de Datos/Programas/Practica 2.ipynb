```python
import re
def limpiar(cad = None):
    """ limpiar
    Función para eliminar caracteres no alfanuméricos de la cadena ingresada usando expresiones regulares
    Params  :
    ---------
    cad : cadena a la que se eliminaran los caracteres no alfanuméricos, también cambia letras con tildes
    por letras sin tildes, (p.e. á=>a, ü=>u, ñ=>n); por omisión es None
    Devuelve:
    ---------
    None si la cadena es None
    res en otro caso, cadena que contiene solamente caracteres alfanuméricos
    conserva los espacios en blanco
    """
    if cad is None:
        return None
    reemplazos = { r"[áä]": "a", r"[éë]": "e", r"[íï]": "i", r"[óö]": "o", r"[úü]": "u",
        r"[ñ]": "n", r"[¿\?\!\¡\,\.]": ""
    }

    cad = cad.lower()
    for exp_reg, remplazo in reemplazos.items():
        cad = re.sub(exp_reg, remplazo, cad)
    return cad

#raise NotImplementedError()
c1 = '¿Mañana vemos una película?'
c2 = '¡Últimamente hemos tenido varias semanas muy lluviosas!'
c1 = limpiar(c1).split() # 'manana vemos una pelicula'.split()
c2 = limpiar(c2).split()
print(c1)
print(c2)
```

```python
def cuenta_letras(cad = None):
    """ cuenta_letras
    Función para contar el número de ocurrencias de cada letra en un texto
    Params  :
    --------
    cad : cadena de la que se obtendrán las cuentas, por omisión es None
    Devuelve:
    ---------
    None si la cadena es None
    res en otro caso, lista de parejas [letra, cuenta] con el número de ocurrencias de cada caracter
    sin contar los espacios en blanco; debe aplicar la función "limpiar" a la cadena de entrada
    para evitar que se contabilicen de forma independiente dos letras que se consideran la misma
    (p.e.: u, ú, ü, ù, û)

    NOTA: La lista de retorno debe estar ordenada de acuerdo a la aparición en la cadena de entrada
    """
    if cad is None:
        return None
    ocurrencias_dict = {}
    ocurrencias_list = []

    for caracter in limpiar(cad).replace(' ', ''):
        if caracter in ocurrencias_dict:
            ocurrencias_dict[caracter] += 1
        else:
            ocurrencias_dict[caracter] = 1

    for caracter, reps in ocurrencias_dict.items():
        ocurrencias_list.append([caracter, reps])
    return ocurrencias_list

#raise NotImplementedError()
c1 = '¡Últimamente hemos tenido varias semanas muy lluviosas!'
cuenta_l = cuenta_letras(c1)
print(cuenta_l)
```

```python
def cuenta_palabras(lista):
    """ cuenta_palabras
Función para contar el número de ocurrencias de cada palabra en una lista de textos
Params  :
---------
  lista : Lista con los textos de los que se obtendrán las cuentas, por omisión es None
Devuelve:
---------
None si la lista es None
 res en otro caso, lista de parejas [palabra, cuenta] con el número de ocurrencias de cada palabra;
        debe aplicar la función "limpiar" a las cadenas de la lista entrada para evitar que se contabilicen
        de forma independiente dos palabras que se consideran la misma (p.e.: qué, Que, ¿¡QUÉ!?)

NOTA: La lista de retorno debe estar ordenada de acuerdo a la aparición en la lista de entrada
"""
    if lista is None:
        return None

    conteo = {}
    orden = []

    for texto in lista:

        texto_limpio = limpiar(texto)

        palabras = texto_limpio.split()
        for palabra in palabras:
            if palabra not in conteo:
                conteo[palabra] = 1
                orden.append(palabra)
            else:
                conteo[palabra] += 1


    resultado = [[palabra, conteo[palabra]] for palabra in orden]
    return resultado

lista = ['Hola, cómo estás!',
         'gana dinero, gana desde casa.',
         'Hola, ¿puedes llamar a casa ahora?',
         'Hola, ¿te puedo llamar mañana?']

print(cuenta_palabras(lista))
cuenta_p = cuenta_palabras(lista)
```
