```python
class Forma(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def mover(self, inc_x=0, inc_y=0):
        self.x += inc_x
        self.y += inc_y
            
class Circulo(Forma):
    def __init__(self, radio = 1, x=0, y=0):
        self.radio = radio
        super().__init__(x, y)
class Cuadrado(Forma): #hereda de object por omisión
    def __init__(self, lado=1, x=0, y=0):
        self.lado = lado
        super().__init__(x, y)
```

```python
ci = Circulo()
print(ci.radio, ci.x, ci.y)
ci.mover(5,7)
print(ci.radio, ci.x, ci.y)
```
