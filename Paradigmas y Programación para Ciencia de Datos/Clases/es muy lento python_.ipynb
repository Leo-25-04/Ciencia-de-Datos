```python
#es muy lento python?
import numpy as np
import numpy.random as rnd
from time import time
from operator import matmul
```

```python
m1 = rnd.rand(300,400)
m2 = rnd.rand(400,200)
```

```python
def my_mm(m1,m2):
    if m1.shape[1] != m2.shape[0]:
        return None
    mr = np.zeros([ m1.shape[0], m2.shape[0] ])
    for i in range(m2.shape[1]):
        for j in range(m2.shape[1]):
            for k in range(mr.shape[0]):
                mr[i][j] += m1[i][k] * m2[k][j]
    return mr
```

```python
def my_time(f, m1, m2):
    t_ini = time()
    f(m1,m2)
    return time()-t_ini
```

```python
print("Numpy:", my_time(np.dot,m1 ,m2))
print("Matmul:", my_time(matmul,m1 ,m2))
print("Python:", my_time(my_mm,m1 ,m2))
```
