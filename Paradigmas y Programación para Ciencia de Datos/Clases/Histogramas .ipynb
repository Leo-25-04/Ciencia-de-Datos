```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
brecha = pd.read_csv ("https://bit.ly/2cLzoxH")
brecha.tail(3)
```

```python
# Histogramas con Pandas
brecha.lifeExp.hist(bins=100)
```

```python
brecha.lifeExp.hist(bins=100, grid = False)
plt.title("Histograma a la Pnadas", fontsize =15)
plt.xlabel("Esperanza de vida", fontsize = 11)
plt.ylabel("Frecuencia", fontsize = 11)
```

```python
brecha.gdpPercap.hist(bins= 500, grid = False)
plt.xlabel("Ingeso Per Cpita")
plt.xscale ("log")
```

```python
# Seaborn
import seaborn as sns
```

```python
sns.displot(brecha.lifeExp, kde= True)
```

```python
df = brecha[brecha.continent == "Africa"]
sns.distplot(df.lifeExp, label = "Africa")
df = brecha[brecha.continent == "Americas"]
sns.distplot(df.lifeExp, label = "America")
```

```python
df = brecha[brecha.continent == "Africa"]
sns.displot(df.lifeExp, label = "Africa")
df = brecha[brecha.continent == "Americas"]
sns.displot(df.lifeExp, label = "America")
```

```python
df = brecha[brecha.continent == "Africa"]
sns.histplot(df.lifeExp, label = "Africa", kde = True, ec="none")
df = brecha[brecha.continent == "Americas"]
sns.histplot(df.lifeExp, label = "America", kde = True, ec="none")
plt.legend()
```

```python
#Histogramas con Matplot.lib
data = np.random.randn(10000)
plt.hist(data, bins = 50)
```

```python
# Sperponer Histogramas com mpl
x1 = np.random.normal(0, .7 , 1000)
x2 = np.random.normal(-2, 1.1, 1000)
x3 = np.random.normal(3, 1.6 , 1000)
#Formato
dic = dict(histtype = "stepfilled",  alpha = .3, bins = 30)

plt.hist(x1, **dic)
plt.hist(x2, **dic)
plt.hist(x3, **dic)
plt.show()
```

```python
num, bordes = np.histogram(data, bins=10)
num, bordes
```

```python
#Histogramas 2D
media = [0,0]
cov = [[1,1],[1,2]]
x, y = np.random.multivariate_normal(media, cov, 10000).T

plt.hist2d(x, y, bins = 50, cmap = "Blues")
cb = plt.colorbar()
cb.set_label("total por contenedor")
```

```python
plt.hexbin(x, y, gridsize = 50, cmap = "Blues")
cb = plt.colorbar()
cb.set_label("Total por contenedor")
```
