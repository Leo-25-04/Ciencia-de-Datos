# BoW: Bag of Words
---

```python
import pandas as pd
import numpy as np
```

```python
# textos a usar
documents = ['Hello , how are you!',
             'Win money, win from home.',
             'Call me now.',
             'Hello, Call hello you tomorrow?']
```

```python
# objeto CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
cv.fit(documents)
nombres = cv.get_feature_names_out()
nombres
```

```python
doc_arr = cv.transform(documents).toarray()
doc_arr
```

```python
freq_mat = pd.DataFrame(data = doc_arr, columns = nombres)
freq_mat
```

## Bayesiano multinomial (inocente, independencia en columnas)

```python
df = pd.read_csv('spam.csv', names  = ['label', 'message'])
df.head(3)
```

```python
# Preprocesamiento
df.label = df.label.map({'ham': 0, 'spam': 1})
df.head()
```

```python
# Entrenaiento/pruebas 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.message, df.label, random_state = 42)
df.shape, X_train.shape, X_test.shape
```

```python
# BoW
cv = CountVectorizer()
train_ = cv.fit_transform(X_train) # Es posible que existan palabras en el de prueba que no estén en el de entrenamiento, en cuyocaso, simplemente no serán consideradas
test_ = cv.transform(X_test)
train_.shape, test_.shape
```

```python
# NaiveBayes
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(train_, y_train)
```

```python
y_pred = nb.predict(test_)
# Evaluación
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score 
print(' Accuracy = ', accuracy_score(y_test, y_pred))
print('Precision = ', precision_score(y_test, y_pred))
print('   Recall = ', recall_score(y_test, y_pred))
print('       F1 = ', f1_score(y_test, y_pred))
```
