```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.2, 0.6, 0.6, 1.5, 1.8])
y = np.array([0.1, 0.4, 0.9, 1.2, 1.7])
plt.scatter(x, y)
plt.show()
```

```python
X_train, y_train = np.array([0.6, 1.8]), np.array([0.4, 1.7])
X_test, y_test = np.array([0.2, 0.6, 1.5]), np.array([0.1, 0.9, 1.2])
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
X_train.resize(len(X_train), 1)
y_train.resize(len(y_train), 1)
lr = LinearRegression()
lr.fit(X_train, y_train)
X_train.shape
```

```python
coefs = lr.coef_[0]
intercept = lr.intercept_[0]
print('y = {:.4f} + {:.4f}x'.format(intercept, coefs[0]))
lr.intercept_, lr.coef_
```

```python
# Visualizaci贸n
xx = np.linspace(0,2,2)
yy = lr.predict(xx.reshape(len(xx), 1))
plt.scatter(x, y)
plt.plot(xx, yy, 'r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error as mse

mse1 = mse(y_test, lr.predict(X_test.reshape(-1, 1)))

print('Error: ', mse1)
```

```python
# Regularizaci贸n Ridge
from sklearn.linear_model import Ridge
rdg = Ridge(alpha = 0.3)
rdg.fit(X_train, y_train) # Ya se les aplic贸 resize por lo que ya tienen la forma necesaria.
coefs = rdg.coef_[0]
intercept = rdg.intercept_[0]
print('y = {:.4f} + {:.4f}x'.format(intercept, coefs))
```

```python
xx = np.linspace(0, 2, 2)
yy = rdg.predict(xx.reshape(-1, 1))
plt.scatter(x, y)
plt.plot(xx, yy, 'g')
```

```python
mse2 = mse(y_test, rdg.predict(X_test.reshape(-1, 1)))

print('Error: ', mse2)
```
